```python
import csv
import os
def format_data(fn):
    rows = list(csv.reader(open(fn)))
    data = [ (int(x), float(y)) for x,y in rows ]
    xs = [ d[0] for d in data ]
    ys = [ d[1] for d in data ]
    return xs,ys

def plot(fn):
    plt.close()
    xs,ys = format_data(fn)
    plt.scatter(xs, ys, s=1)
```

```python
plot('times_initial_wps.csv')
```

```python
plot('times_prealloc_wps.csv')
```

```python
plot('times_hier_wps.csv')
```

```python
plt.close()
fig, (a1, a2, a3) = plt.subplots(
    1, 3, sharex=True, sharey=True,
    num=1, figsize=(12,6))
xs,ys = format_data('times_initial_wps.csv')
a1.scatter(xs, ys, s=2, c='b', linewidths=0)
a1.set_title('Initial Design')
xlim(0, 1600)
ylim(0, 6000)
xticks([0, 400, 800, 1200, 1600])
xs,ys = format_data('times_prealloc_wps.csv')
a2.scatter(xs, ys, s=2, c='b', linewidths=0)
a2.set_title('With Preallocation')
xs,ys = format_data('times_hier_wps.csv')
a3.scatter(xs, ys, s=2, c='b', linewidths=0)
a3.set_title('With Hierarchical Minutes')
fig
```
