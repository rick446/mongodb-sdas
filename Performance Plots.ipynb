```python
import csv
import os
def format_data(fn):
    rows = list(csv.reader(open(fn)))
    data = [ (int(x), float(y)) for x,y in rows ]
    xs = [ d[0] for d in data ]
    ys = [ d[1] for d in data ]
    return xs,ys

def plot(fn):
    xs,ys = format_data(fn)
    plt.scatter(xs, ys, s=1)
    
def plot_on_axis(title, fn, axis):
    xs, ys = format_data(fn)
    axis.scatter(xs, ys, s=2, c='b', linewidths=0)
    axis.set_title(title)
    axis.set_xlim(0, 1440)
    axis.set_ylim(0, 6000)
    axis.set_xticks([0, 360, 720, 1080, 1440])
    axis.set_xticklabels(
           ['00:00', '06:00', '12:00', '18:00', '00:00'])
    
plt.close()
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(1, 3, sharex=False, sharey=True, num=1, figsize=(12,5))
plot_on_axis('Initial Design', 'times-initial.csv', ax0)
plot_on_axis('With Preallocation', 'times-prealloc.csv', ax1)
plot_on_axis('With Hierarchical Minutes', 'times-hier.csv', ax2)
fig.savefig('Combined.png')
fig
```
